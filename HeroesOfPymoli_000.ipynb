```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file)
purchase_data_df.head()
```

```python
# Check data for rows with null cells
purchase_data_df.count()
```

## Player Count

* Display the total number of players


```python
#Player count is the numbers of uniqe SN
SN_unique = len(purchase_data_df["SN"].unique())
total_players_df = pd.DataFrame({"Total Players":[SN_unique]})

total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Unique no of items
item_count=len(purchase_data_df["Item Name"].unique())

#Average price of purchased items
average_price=purchase_data_df["Price"].mean()

# Number of purchases
purchase_count=len(purchase_data_df["Purchase ID"].unique())

#Total sales
total_sales =purchase_data_df["Price"].sum()

#item_count
#average_price
#purchase_count
#total_sales

#Summary in a dataFrame
summary_df =pd.DataFrame({"Number of Unique Item":[item_count],
                         "Average Price": average_price,
                         "Number of Purchases" : purchase_count,
                         "Total Revenue": total_sales})
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:.2f}".format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Get a dataframe with no duplicates of SN
unique_SN_df= purchase_data_df.drop_duplicates(subset=["SN"])

# The computations on of percentage and counts will be on the dataframe with uniqe players
gender_count = unique_SN_df["Gender"].value_counts()
gender_count_df = pd.DataFrame({"Total Unique Count":gender_count})
percentage_of_players= gender_count_df["Total Unique Count"]/gender_count_df["Total Unique Count"].sum()

#Add percentage_of_players to the gender_count_df
gender_count_df["Percentage of Players"] =percentage_of_players

#Format the Percentage of Players column
gender_count_df["Percentage of Players"]=gender_count_df["Percentage of Players"].map("{:,.2%}".format)

gender_count_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_grp_df =purchase_data_df.groupby("Gender").Price.agg(["count", "mean","sum"])
purchase_grp_df
```

```python
#Compute the count, mean and sum of hte purchase group
purchase_grp_df =purchase_data_df.groupby("Gender").Price.agg(["count", "mean","sum"])

#Obtaiin the unique gender count
purchase_grp_df["Uniq_gender_count"] = unique_SN_df["Gender"].value_counts()

purchase_grp_df["Avg Total Purchase per Person"] = (purchase_grp_df["sum"]/purchase_grp_df["Uniq_gender_count"]).map("${:.2f}".format)
#del b_unique_item_df ["Age Range", "Gender", "Purchase ID", "Age"]
del purchase_grp_df["Uniq_gender_count"]
purchase_grp_df=purchase_grp_df.rename(columns = {"count":"Purchase Count","mean":"Average Purchase Price", "sum":"Total Purchase Value"})
#format columns
purchase_grp_df["Average Purchase Price"]=purchase_grp_df["Average Purchase Price"].map("${:.2f}".format)
purchase_grp_df["Total Purchase Value"]=purchase_grp_df["Total Purchase Value"].map("${:.2f}".format)

purchase_grp_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Define Bins
bins =[0, 9, 14, 19, 24, 29, 34, 40,100 ]
```

```python
#Group names
bin_names=["<10", "10 to 14", "15 to 19", "20 to 24", "25 to 29", "30 to 34", "35 to 39", "40+"]
```

```python
#pd.cut(unique_SN_df["Age"], bins, labels=bin_names, include_lowest=True)

unique_SN_df["Age Range"] = pd.cut(unique_SN_df["Age"], bins, labels=bin_names, include_lowest=True)

#df["Test Score Summary"] = pd.cut(df["Test Score"], bins, labels=group_names, include_lowest=True)
unique_SN_df.head()
```

```python
#Create groupby based on Age range
Age_Ranges= unique_SN_df.groupby("Age Range")
print(Age_Ranges["Age"].mean())
```

```python
# get the count of each range
counts_in_Age_Ranges=Age_Ranges["Age"].count()
counts_in_Age_Ranges_df =  pd.DataFrame({"Total Range Count":counts_in_Age_Ranges})
percente_of_players= counts_in_Age_Ranges_df["Total Range Count"]/counts_in_Age_Ranges_df["Total Range Count"].sum()

#Add percentage_of_players to the counts_in_Age_Ranges_df
counts_in_Age_Ranges_df["Players % In Range"] =percente_of_players

#Format the Percentage of Players column
counts_in_Age_Ranges_df["Players % In Range"]=counts_in_Age_Ranges_df["Players % In Range"].map("{:,.2%}".format)
counts_in_Age_Ranges_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
# Define Bins
bins =[0, 9, 14, 19, 24, 29, 34, 40,100 ]

#Group names
bin_names=["<10", "10 to 14", "15 to 19", "20 to 24", "25 to 29", "30 to 34", "35 to 39", "40+"]
```

```python
#pd.cut(unique_SN_df["Age"], bins, labels=bin_names, include_lowest=True)
#pd.cut(purchase_data_df["Age"], bins, labels=bin_names, include_lowest=True)

purchase_data_df["Age Range"] = pd.cut(purchase_data_df["Age"], bins, labels=bin_names, include_lowest=True)

#df["Test Score Summary"] = pd.cut(df["Test Score"], bins, labels=group_names, include_lowest=True)
purchase_data_df.head()
```

```python
Range_of_Ages_df =purchase_data_df.groupby("Age Range").Price.agg(["count", "mean","sum"])
Range_of_Ages_df.rename(columns={"count":"Purchase Count", "mean":"Average Purchase Price", "sum":"Total Purchase Value"}, inplace=True)
Range_of_Ages_df
```

```python
#Add a column that shows the count of unique SN in age ranges
Range_of_Ages_df["Total Range Count"] = counts_in_Age_Ranges_df["Total Range Count"]
# compute theAverage Total purchase per person per age range
Range_of_Ages_df["Avg Total Purchase per Person"] = (Range_of_Ages_df["Total Purchase Value"]/Range_of_Ages_df["Total Range Count"])
#Delete the column that shows the count of unique SN in age ranges as it is not aa required output
del Range_of_Ages_df["Total Range Count"]
#format columns
Range_of_Ages_df["Average Purchase Price"]=Range_of_Ages_df["Average Purchase Price"].map("${:.2f}".format)
Range_of_Ages_df["Total Purchase Value"]=Range_of_Ages_df["Total Purchase Value"].map("${:.2f}".format)
Range_of_Ages_df["Avg Total Purchase per Person"]=Range_of_Ages_df["Avg Total Purchase per Person"].map("${:.2f}".format)
#show the df
Range_of_Ages_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items_df=purchase_data_df[["Item ID", "Item Name", "Price"]]
popular_items_df.head()
```

```python
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
b_unique_item_df = popular_items_df.groupby(["Item ID","Item Name"]).Price.agg(["count", "mean","sum"])
b_unique_item_df=b_unique_item_df.sort_values(["count"], ascending=False)
b_unique_item_df.rename(columns={"count":"Purchase Count", "mean":"Item Price", "sum":"Total Purchase Value"}, inplace=True)

#format
b_unique_item_df["Item Price"]=b_unique_item_df["Item Price"].map("${:.2f}".format)
b_unique_item_df["Total Purchase Value"]=b_unique_item_df["Total Purchase Value"].map("${:.2f}".format)

b_unique_item_df.head()
```
